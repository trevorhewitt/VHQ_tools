## Measuring internal consistency with Cronbach's alpha

### Table 2

```python
# ============================================================
# Cronbach's alpha + bootstrap CI, per dimension and experiment
# ============================================================

# --- Dimension mapping (6 dims × 3 items = 18 items) ---
dim_items = {
    dim: [f"Q{i:02d}" for i in range(1 + 3*k, 1 + 3*k + 3)]
    for k, dim in enumerate(dimension_labels)
}

def cronbach_alpha(df_items: pd.DataFrame) -> float:
    """
    Cronbach's alpha for columns in df_items.
    Rows with any missing values are dropped (listwise for that dimension).
    """
    x = df_items.dropna(axis=0, how="any").to_numpy(dtype=float)
    n, k = x.shape
    if k < 2 or n < 2:
        return np.nan
    item_vars = x.var(axis=0, ddof=1)
    total_var = x.sum(axis=1).var(ddof=1)
    if total_var <= 0:
        return np.nan
    return (k / (k - 1)) * (1 - item_vars.sum() / total_var)

def bootstrap_alpha_ci(df_items: pd.DataFrame, n_boot: int = 5000, seed: int = 0):
    """
    Percentile bootstrap 95% CI for alpha.
    """
    x = df_items.dropna(axis=0, how="any").to_numpy(dtype=float)
    n = x.shape[0]
    if n < 2:
        return (np.nan, np.nan, np.nan)

    rng = np.random.default_rng(seed)
    alphas = np.empty(n_boot, dtype=float)

    for b in range(n_boot):
        idx = rng.integers(0, n, size=n)
        xb = x[idx, :]
        k = xb.shape[1]
        item_vars = xb.var(axis=0, ddof=1)
        total_var = xb.sum(axis=1).var(ddof=1)
        if total_var <= 0 or k < 2:
            alphas[b] = np.nan
        else:
            alphas[b] = (k / (k - 1)) * (1 - item_vars.sum() / total_var)

    alphas = alphas[~np.isnan(alphas)]
    if alphas.size == 0:
        return (np.nan, np.nan, np.nan)

    lo, hi = np.percentile(alphas, [2.5, 97.5])
    return (np.nanmean(alphas), lo, hi)

def fmt_alpha(x):
    if pd.isna(x):
        return ""
    s = f"{x:.2f}"
    return s[1:] if s.startswith("0") else s

# --- Experiment definitions ---
exp_defs = [
    (1, "Images", df_vhq_img_items),
    (2, "SIVHs", df_vhq_lab_items),
    (3, "Psychedelics", df_vhq_rec_items),
]

# --- Precompute Ns (dimension-wise listwise N can vary; headline N uses any non-missing row across Q01-Q18) ---
q_cols = [f"Q{i:02d}" for i in range(1, 19)]

def n_responses_any(df):
    # counts rows with at least one non-missing response among Q01..Q18
    return int(df[q_cols].notna().any(axis=1).sum())

exp_headers = {}
for num, name, df in exp_defs:
    N = n_responses_any(df)
    exp_headers[num] = f"Experiment {num} ({name}; N={N})"

# --- Pooled data (row-bind, keep only Q01..Q18) ---
df_pooled = pd.concat([df[q_cols] for _, _, df in exp_defs], ignore_index=True)
N_pooled = n_responses_any(df_pooled)
pooled_header = f"Pooled data (N={N_pooled})"

# --- Compute alphas and pooled CIs (CI only for pooled) ---
rows = []
n_boot = 5000     
seed = 0

for dim in dimension_labels:
    cols = dim_items[dim]

    row = {"Dimension": dim}

    # experiments 1-3
    for num, _, df in exp_defs:
        a = cronbach_alpha(df[cols])
        row[exp_headers[num]] = a

    # pooled alpha + CI
    pooled_mean, pooled_lo, pooled_hi = bootstrap_alpha_ci(df_pooled[cols], n_boot=n_boot, seed=seed)
    row[pooled_header] = cronbach_alpha(df_pooled[cols])  # report alpha on observed pooled data
    row["95% CI (lower; pooled)"] = pooled_lo
    row["95% CI (upper; pooled)"] = pooled_hi

    rows.append(row)

out = pd.DataFrame(rows)

# --- Format alphas/CI to two decimals like .00 ---
for c in out.columns:
    if c != "Dimension":
        out[c] = out[c].apply(fmt_alpha)

out
```

## 11D-ASC Internal consistency
### Table 3

```python
# ============================================
# Cronbach's alpha for 11D-ASC (dimension-wise)
# ============================================

def cronbach_alpha(df_items: pd.DataFrame) -> float:
    """
    Cronbach's alpha with listwise deletion.
    """
    x = df_items.dropna(axis=0, how="any").to_numpy(dtype=float)
    n, k = x.shape
    if k < 2 or n < 2:
        return np.nan
    item_vars = x.var(axis=0, ddof=1)
    total_var = x.sum(axis=1).var(ddof=1)
    if total_var <= 0:
        return np.nan
    return (k / (k - 1)) * (1 - item_vars.sum() / total_var)

def fmt_alpha(x):
    if pd.isna(x):
        return ""
    s = f"{x:.2f}"
    return s[1:] if s.startswith("0") else s

# --- Ensure question numbers map cleanly to Qxx columns ---
asc_key = asc_question_key.copy()
asc_key["Qcol"] = asc_key["question"].apply(lambda q: f"Q{int(q):02d}")

# --- Compute alpha per dimension, preserving key order ---
rows = []

for dim_name, sub in asc_key.groupby("name", sort=False):
    cols = sub["Qcol"].tolist()
    cols = [c for c in cols if c in df_asc_rec_items.columns]  # safety check
    alpha = cronbach_alpha(df_asc_rec_items[cols])
    rows.append({
        "Dimension": dim_name,
        "Cronbach’s α": alpha
    })

out_asc_alpha = pd.DataFrame(rows)
out_asc_alpha["Cronbach’s α"] = out_asc_alpha["Cronbach’s α"].apply(fmt_alpha)

display(out_asc_alpha)
```

## Confirmatory Factor Analysis

### Table 5

```python
# ============================================
# 6-factor factor analysis on pooled VHQ items
# ============================================

# --- Keep only questionnaire items ---
q_cols = [f"Q{i:02d}" for i in range(1, 19)]

# --- Pool data (row-bind) ---
df_pooled = pd.concat(
    [df[q_cols] for df in [df_vhq_img_items, df_vhq_lab_items, df_vhq_rec_items]],
    ignore_index=True
)

# --- Listwise deletion for FA ---
df_fa = df_pooled.dropna(axis=0, how="any")

# --- Fit 6-factor model ---
fa = FactorAnalyzer(
    n_factors=6,
    method="ml",        # maximum likelihood
    rotation="varimax"  # orthogonal rotation
)
fa.fit(df_fa)

# --- Factor loadings ---
loadings = pd.DataFrame(
    fa.loadings_,
    index=q_cols,
    columns=[f"Factor {i+1}" for i in range(6)]
)

# --- variance explained ---
variance = pd.DataFrame(
    {
        "SS Loadings": fa.get_factor_variance()[0],
        "Proportion Var": fa.get_factor_variance()[1],
        "Cumulative Var": fa.get_factor_variance()[2],
    },
    index=[f"Factor {i+1}" for i in range(6)]
)


table_loadings = (
    loadings
    .loc[[f"Q{i:02d}" for i in range(1, 19)]]
    .round(2)
    .reset_index()
    .rename(columns={"index": "Item"})
)

table_loadings
```

## Correlation Matrix

### Table 4

```python
# ============================================
# Item–item Pearson correlation matrix (pooled)
# ============================================

# --- Keep only questionnaire items ---
q_cols = [f"Q{i:02d}" for i in range(1, 19)]

# --- Pool data ---
df_pooled = pd.concat(
    [df[q_cols] for df in [df_vhq_img_items, df_vhq_lab_items, df_vhq_rec_items]],
    ignore_index=True
)

# --- Listwise deletion ---
df_corr = df_pooled.dropna(axis=0, how="any")

# --- Pearson correlation matrix ---
corr = df_corr.corr(method="pearson")

# --- Formatting helpers ---
def fmt_r(x):
    if pd.isna(x):
        return ""
    s = f"{x:.2f}"
    return s[1:] if s.startswith("0") else s

# --- table ---
corr_table = corr.copy()

# Optionally blank the diagonal (common in papers)
np.fill_diagonal(corr_table.values, np.nan)

corr_table = corr_table.applymap(fmt_r)

corr_table
```
